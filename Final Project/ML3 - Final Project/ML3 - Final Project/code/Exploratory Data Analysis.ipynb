```python
def check_feature_value(dataframe):
    
    feature_list = list(dataframe)
    for index, feature in enumerate(feature_list):
        print('\n\nIndex: ',index)
        print('Feature: ',feature)
        print('Unique Elements: ',dataframe[feature].nunique(dropna=False))
        print('Values of each elements:\n',dataframe[feature].value_counts(dropna=False))
        #brk = input('Press Enter for next feature details!')
```
